# preprocess

```python
def preprocess(text):
    # Step 1 — split sentence into individual words
    tokens = word_tokenize(str(text).lower())
    
    # Step 2 — remove stopwords and non-alphabet words
    filtered = [word for word in tokens if word.isalpha() and word not in stop_words]
    
    # Step 3 — lemmatize (reduce words to root form)
    lemmatised = {lemm.lemmatize(word) for word in filtered}
    
    return lemmatised
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
lemm = WordNetLemmatizer()
```

# Intent Classifier

```python
INTENT_TEMPLATES = {
    "predict_winner":    "will win beat predict winner future",
    "head_to_head":      "how many times won beaten total count",
    "home_wins":         "home won hosting ground home record",
    "away_wins":         "away won travel visiting away record",
    "draws":             "draw drew drawn tied tie",
    "last_match":        "last recent latest result score",
    "best_tournament":   "best tournament competition perform wins",
    "worst_performance": "worst loss embarrassing defeat humiliated terrible",
    "win_streak":        "longest streak consecutive wins row kept winning",
    "lose_streak":       "longest losing streak consecutive losses row kept losing",
    "highest_scoring":   "most goals highest scoring thrilling offensive attack",
    "biggest_win":       "biggest dominant victory thrashing hammered crushed margin",
}
```

```python
def classify_intent(query):
    query_tokens = preprocess(query)
    best_intent, best_score = None, -1
    
    for intent, template in INTENT_TEMPLATES.items():
        template_tokens = preprocess(template)
        overlap = len(query_tokens.intersection(template_tokens))
        if overlap > best_score:
            best_score = overlap
            best_intent = intent
    
    # keyword override — only REPLACES if a keyword is found
    q_lower = query.lower()
    if any(w in q_lower for w in ["will", "who wins", "going to win"]):
        return "predict_winner"
    elif any(w in q_lower for w in ["draw", "drew", "drawn"]):
        return "draws"
    elif "home" in q_lower:
        return "home_wins"
    elif "away" in q_lower:
        return "away_wins"
    elif any(w in q_lower for w in ["last", "recent", "latest"]):
        return "last_match"
    elif any(w in q_lower for w in ["worst", "biggest loss", "embarrass"]):
        return "worst_performance"
    elif any(w in q_lower for w in ["win streak", "winning streak", "consecutive wins"]):
        return "win_streak"
    elif any(w in q_lower for w in ["lose streak", "losing streak", "consecutive loss"]):
        return "lose_streak"
    elif any(w in q_lower for w in ["highest scoring", "most goals"]):
        return "highest_scoring"
    elif any(w in q_lower for w in ["biggest win", "biggest victory", "most dominant"]):
        return "biggest_win"
    # venue based questions also mean prediction
    elif any(w in q_lower for w in [" in ", " at ", "hosted by", "held in"]):
        return "predict_winner"   # "will brazil win IN argentina?" = prediction
    
    return best_intent
    # if NO keyword matched, return whatever the loop found  
```

```python
#testing intent classifier
print(classify_intent("Will Brazil beat Argentina?"))
print(classify_intent("How many times has Germany beaten France?"))
print(classify_intent("How many draws between Spain and Italy?"))
print(classify_intent("What is Brazil's longest winning streak?"))
print(classify_intent("What was England's worst performance?"))
```

# Extracting Country


```python
import re
ALL_TEAMS = set(data['home_team'].unique())|set(data['away_team'].unique())
def extract_countries(query):
    q_lower = query.lower()
    found = []
    for team in sorted(ALL_TEAMS, key=len, reverse=True):
        if re.search(r'\b' + re.escape(team.lower()) + r'\b', q_lower):
            found.append(team)
            if len(found) == 2:
                break
    found.sort(key=lambda t: q_lower.index(t.lower()))

    # ── detect venue ──────────────────────────────────────────
    venue = None
    venue_keywords = [" in ", " at ", " hosted by ", " held in ", " playing in "]
    for keyword in venue_keywords:
        if keyword in q_lower:
            keyword_position = q_lower.index(keyword)
            for team in sorted(ALL_TEAMS, key=len, reverse=True):
                # ✅ start searching ONLY from keyword_position onwards
                team_position = q_lower.find(team.lower(), keyword_position)
                if team_position != -1 and team_position > keyword_position:
                    venue = team
                    break
            break

    return found, venue
```

```python
# Test
print(extract_countries("Will Brazil win against Argentina in Argentina?"))
print(extract_countries("Will France beat England in England?"))
print(extract_countries("Will Brazil beat Argentina?"))
```

# Stat Helper

```python
def get_h2h(t1, t2):
    return data[
        ((data['home_team'] == t1) & (data['away_team'] == t2))|
        ((data['home_team'] == t2) & (data['away_team' == t1]))
        ]

def team_won(row, team):
    if row['home_team'] == team:
        return row['home_score'] > row['away_score']
    return row['home_score'] < row['away_score']
```

```python
#get_h2h It returns a filtered dataframe, just the rows where these two teams played each other
#team_won tells us if the home team won or not, if home score is more than away score it returns true, if away score is more it returns false
```
